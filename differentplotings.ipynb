```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y,marker='o',linestyle='--',color='b')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({'x':[1,2,3,4],'y':[10,20,25,30]})
sns.lineplot(x='x',y='y',data=df,marker='o')
plt.title('Line Plot')
plt.show()
```

```python
import plotly.express as px
df=pd.DataFrame({'x':[1,2,3,4],'y':[10,20,25,30]})
fig=px.line(df,x='x',y='y',markers=True)
fig.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,25,30]
plt.scatter(x,y,color='r',s=50)
plt.title('Scatter Plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({'x':[1,2,3,4],'y':[10,20,25,30]})
sns.scatterplot(x='x',y='y',data=df)
plt.title('Scatter Plot')
plt.show()
```

```python
import plotly.express as px
df=pd.DataFrame({'x':[1,2,3,4],'y':[10,20,25,30]})
fig=px.scatter(df,x='x',y='y',title='Scatter Plot')
fig.show()
```

```python
import matplotlib.pyplot as plt
categories=['A','B','C']
values=[4,3,5]
plt.bar(categories,values,color='g')
plt.title('Bar Chart')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({'categories':['A','B','C'],'values':[4,3,5]})
sns.barplot(x='categories',y='values',data=df)
plt.title('Bar Chart')
plt.show()
```

```python
import plotly.express as px
df=pd.DataFrame({'categories':['A','B','C'],'values':[4,3,5]})
fig=px.bar(df,x='categories',y='values',title='Bar Chart')
fig.show()
```

```python
import matplotlib.pyplot as plt
data=[1,2,2,3,3,3,4]
plt.hist(data,bins=4,color='c',edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=[1,2,2,3,3,3,4]
sns.histplot(data,bins=4,kde=True)
plt.title('Histogram')
plt.show()
```

```python
import plotly.express as px
data=[1,2,2,3,3,3,4]
fig=px.histogram(data,nbins=4,title='Histogram')
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.rand(5,5)
plt.imshow(data,cmap='viridis')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data=np.random.rand(5,5)
sns.heatmap(data,cmap='viridis')
plt.title('Heatmap')
plt.show()
```

```python
import plotly.express as px
import numpy as np
data=np.random.rand(5,5)
fig=px.imshow(data,color_continuous_scale='viridis',title='Heatmap')
fig.show()
```

```python
import matplotlib.pyplot as plt
data=[[1,2,3],[4,5,6],[7,8,9]]
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({'Group':['A']*3+['B']*3+['C']*3,'Value':[1,2,3,4,5,6,7,8,9]})
sns.boxplot(x='Group',y='Value',data=df)
plt.title('Box Plot')
plt.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({'Group':['A']*3+['B']*3+['C']*3,'Value':[1,2,3,4,5,6,7,8,9]})
fig=px.box(df,x='Group',y='Value',title='Box Plot')
fig.show()
```
