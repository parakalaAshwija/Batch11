```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(10)
blood_sugar = np.concatenate([
    np.random.normal(95, 10, 120),   # normal group
    np.random.normal(140, 15, 60),   # pre-diabetic group
    np.random.normal(180, 20, 20)    # diabetic group
])

# Step 2: Create KDE Plot
plt.figure(figsize=(10,6))
sns.kdeplot(blood_sugar, fill=True, color='skyblue', linewidth=2)

# Step 3: Add labels and title
plt.title("Density Plot of Blood Sugar Levels (KDE Plot)", fontsize=16)
plt.xlabel("Blood Sugar Level (mg/dL)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

# Step 4: Display plot
plt.show()
```
